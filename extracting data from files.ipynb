```python
import unicodedata
import PyPDF2
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Ensure required resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Read the PDF
pdfreader = PyPDF2.PdfReader("file1pdf.pdf")  # Replace with your actual PDF file
pageObj = pdfreader.pages[2]  
page2 = pageObj.extract_text()

# Define punctuation and stop words
punctuations = ['(', ')', ';', ':', '[', ']', ',', '...', '.']
tokens = word_tokenize(page2)
stop_words = set(stopwords.words('english'))

# Normalize and filter words
keywords = [
    unicodedata.normalize("NFKC", word) 
    for word in tokens 
    if word.lower() not in stop_words and word not in punctuations
]

print(keywords)
```

```python
name_list = list()
check = ['Mr.','Mrs.','Ms.']
for idx, token in enumerate(tokens):
    if token.startswith(tuple(check)) and idx < (len(tokens)-1):
        name = token + tokens[idx+1] + ' ' + tokens[idx+2]
        name_list.append(name)
    print(name_list)
```

```python
import unicodedata
import urllib.request

# Function to remove non-printable characters
def clean_text(text):
    return "".join(ch for ch in text if unicodedata.category(ch)[0] != "C")

# Fetch webpage content
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
response = urllib.request.urlopen(url)
html_doc = response.read().decode('utf-8')

# Clean any hidden characters in the response
html_doc = clean_text(html_doc)

# Print first 500 characters as a preview
print(html_doc[:500])
```
